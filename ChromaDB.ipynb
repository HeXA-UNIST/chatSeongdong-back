```python
import os
import json
import chromadb

def read_jsonl_files(directory):
    documents = []
    for file in os.listdir(directory):
        if file.endswith('.jsonl'):
            with open(os.path.join(directory, file), 'r') as f:
                for line in f:
                    obj = json.loads(line)
                    messages = obj['messages']
                    if messages:
                        first_message_content = messages[0]['content']
                        for message in messages:
                            documents.append((message['content'], first_message_content))
    return documents

client = chromadb.Client()
collection = None
try:
  collection = client.create_collection("jsonl_messages")
except:
  collection = client.get_collection("jsonl_messages")
data_directory = 'data'
documents = read_jsonl_files(data_directory)

# Initialize a counter for the document ID
doc_id = 0

for content, first_message_content in documents:
    collection.add(
        documents=[content],
        metadatas=[{"source": first_message_content}],
        ids=[str(doc_id)]  # Convert the integer ID to a string
    )
    doc_id += 1  # Increment the ID for the next document
```
